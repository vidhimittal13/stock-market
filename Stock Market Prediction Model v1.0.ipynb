```python
sp500["Tomorrow"] = sp500["Close"].shift(-1) 
```

```python
#want to set up target based on tomorrow's price
#converting to integer so that we can use it in machine learning
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
#if you go back too far, the market couldve shifted fundamentally
#deleting everything before 1990
```

```python
sp500 = sp500.loc["1990-01-01":].copy() #copy because want to avoid error of subsetting and then sign back
```

```python
sp500 #only dates after january 1st 1990
```

```python
from sklearn.ensemble import RandomForestClassifier
#trains a bunch of individual decision trees with randomized parameters
#and then averaging the results
#resistent to overfit
#run relatively quickly
#can pick up non-linear tendencies such as open price

#n_estimators is the number of decision trees we want to train
#min samples split protects against overfitting if they build too much....find optimal number
#random state--- if we run the same model twice, the number will be in the predictable sequence
# split into train and set...this is time series data...cannot use cross validation bc horrible in real world
#bc if u dont take time series nature into account, u will use future data to predict the past and will lead to leakage
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split = 100, random_state = 1)

#everything except last 100 rows in training set
#last 100 rows in test set
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

#Don't use tomorrow or target
predictors = ["Close","Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
#measure how accurate the model is
#precision score --- what percentage of times we said market would go up did it actually go up
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index = test.index)
```

```python
preds
```

```python
precision_score(test["Target"], preds)
```

```python
#plotting our predictions...concatenating test and target
combined = pd.concat([test["Target"], preds], axis = 1)
```

```python
combined.plot()
```

```python
#building a more robust way

#test accross multiple years

#back testing

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis =1)
    return combined
```

```python
# 10 years, training model for about a year then going to next year

def backtest(data, model, predictors, start = 2500, step = 250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts() #how many times each was made
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
#more predictors

horizons = [2,5,60,250,1000] #mean close price in last 2 days, last week, last 3 months, last year, last 4 years

#may be due for upside/downside

new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+= [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
sp500
```

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state = 1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds>= .6] = 1
    preds[preds < .6] = 0 #reduces total number of trading days
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis =1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
